# Activating Your First Neural Network

```python
# Initial imports
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs, make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

%matplotlib inline
```

```python
# Import Sequential and Dense from Tensorflow Keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

## Using a Single Neuron for Binary Classification

### Creating Dummy Data

```python
# Generate 1000 demo data samples with 2 features and two centers

X, y = make_blobs(n_samples=1000, centers=2, n_features=2, random_state=1)
```

```python
y = y.reshape(-1, 1)
```

```python
# Creating a DataFrame with the dummy data
df = pd.DataFrame(X, columns=["Feature 1", "Feature 2"])
df["Target"] = y
```

```python
# Plot the dummy data
df.plot.scatter(x="Feature 1", y="Feature 2", c="Target", colormap="winter")
```

### Data Preprocessing

Use `train_test_split` to create training and testing data.

```python
# Create training and testing datasets

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

Scale the data using the `StandarScaler`.

```python
# Create scaler instance
X_scaler = StandardScaler()
```

```python
# Fit the scaler
X_scaler.fit(X_train)
```

```python
# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test) 
```

### Neural Network Model Creation

```python
# Create a sequential model
neuron = Sequential()
```

#### Add the first layer of our neural network.

```python
# First layer
number_inputs = 2
```

#### We finish creating our neural network by adding the output layer.


```python
# Output layer
number_hidden_nodes = 3
```

```python
# Model summary
neuron.add(Dense(units=number_hidden_nodes, activation="relu", input_dim=number_inputs))
```

```python
number_classes = 1

neuron.add(Dense(units=number_classes, activation="sigmoid"))
```

```python
# Display model summary
neuron.summary()
```
